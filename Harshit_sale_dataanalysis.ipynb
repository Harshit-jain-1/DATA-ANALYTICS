```python
# Filter for sales in September
september_sales = df[df['Date'].dt.month == 9]

# Group by product and sum the sales
sales_summary = september_sales.groupby('Item Name')['Primary Sales'].sum().reset_index()

# Find the highest-selling product
highest_selling_product = sales_summary.loc[sales_summary['Primary Sales'].idxmax()]

print("\nHighest-selling product in September:")
print(highest_selling_product)
```

```python
# Filter for the "CND Chennai" sales team and sales in May
may_sales_cnd_chennai = df[(df['Sales Team'] == 'CND Chennai') & (df['Date'].dt.month == 5)]

# Group by product and sum the sales
sales_summary = may_sales_cnd_chennai.groupby('Item Name')['Primary Sales'].sum().reset_index()

# Find the highest-selling product
highest_selling_product = sales_summary.loc[sales_summary['Primary Sales'].idxmax()]

print("\nHighest-selling product for 'CND Chennai' in May:")
print(highest_selling_product)
```

```python
# Filter for Bangalore HQ and June
june_data_bangalore = df[(df['HQ'] == 'HQ-Bangalore') & (df['Date'].dt.month == 6)]

# Group by customer and sum the stock returns
returns_summary = june_data_bangalore.groupby('Customer')['Sales Return'].sum().reset_index()

# Find the customer with the maximum stock return
max_return_customer = returns_summary.loc[returns_summary['Sales Return'].idxmax()]

print("Customer with maximum stock returns in June for Bangalore HQ:")
print(max_return_customer)
```

```python
# Group by sales team and sum the primary sales and returned sales
sales_summary = df.groupby('Sales Team').agg({
    'Primary Sales': 'sum',
    'Sales Return': 'sum'
}).reset_index()

# Calculate the percentage of returned sales due to expiry
sales_summary['Return_Percentage'] = (sales_summary['Sales Return'] / sales_summary['Primary Sales']) * 100

# Find the sales team with the maximum return percentage
max_return_team = sales_summary.loc[sales_summary['Return_Percentage'].idxmax()]

print("Sales team with the maximum percentage of primary sales returned due to expiry:")
print(max_return_team)
```

```python
# Calculate overall primary sales and total sales affected by breakage
total_primary_sales = df['Primary Sales'].sum()
total_sales_affected_by_breakage = df['Breakage'].sum()

# Calculate the percentage of overall primary sales affected by breakage
percentage_affected_by_breakage = (total_sales_affected_by_breakage / total_primary_sales) * 100

print(f"Percentage of overall primary sales affected by breakage: {percentage_affected_by_breakage:.2f}%")
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Set the Date as the index 
df4_num_mode.set_index('Date', inplace=True)

# Fit an ARIMA model (p, d, q) - parameters can be adjusted based on data characteristics
model = ARIMA(df4_num_mode['Primary Sales'], order=(1, 1, 1))
model_fit = model.fit()

# Forecast for October (1 step ahead)
forecast = model_fit.forecast(steps=1)

# Display the forecasted value
print(f"Forecasted primary sales for October: {forecast}")
```

```python
# Create a DataFrame for forecasted sales for "Britorva 20"
forecasted_sales_df = pd.DataFrame({
    'Item Name': ['Britorva 20'],
    'Forecasted Sales': [forecast.values[0]],
    'HQ': ['HQ-Coimbatore'],
    'Date': ['October']
})

# Filter for "Britorva 20" under Coimbatore HQ in October
forecasted_sales_britorva_20 = forecasted_sales_df[(forecasted_sales_df['Item Name'] == 'Britorva 20') & 
                                                    (forecasted_sales_df['HQ'] == 'HQ-Coimbatore') & 
                                                    (forecasted_sales_df['Date'] == 'October')]

# Check if the product was found and display the result
if not forecasted_sales_britorva_20.empty:
    sales_value = forecasted_sales_britorva_20['Forecasted Sales'].values[0]
    print(f"The forecasted sales of 'Britorva 20' in October under Coimbatore HQ is: {sales_value}.")
else:
    print("No forecasted sales data found for 'Britorva 20' in October under Coimbatore HQ.")
```

```python
# Set the Date as the index
df4_num_mode.set_index('Date', inplace=True)

# Prepare a DataFrame to store forecasted sales
forecasted_sales = {}

# Fit an ARIMA model for each product and forecast for November (2 steps ahead)
for product in df4_num_mode.columns:
    model = ARIMA(df4_num_mode['Primary Sales'], order=(1, 1, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=2)  # Forecast for November and December
    forecasted_sales[product] = forecast

# Convert the forecasted sales to a DataFrame
forecasted_df = pd.DataFrame(forecasted_sales)

# Display the forecasted sales for November
print("Forecasted Sales for November:")
print(forecasted_df)

# Find the product with the highest forecasted sales for November
highest_sales_product = forecasted_df.iloc[0].idxmax()
highest_sales_value = forecasted_df.iloc[0].max()

# Display the result
print(f"\nThe product with the highest forecasted sales for 'CND Chennai' in November is: {highest_sales_product} with sales of {highest_sales_value}.")
```
